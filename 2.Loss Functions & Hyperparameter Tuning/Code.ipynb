	Loss Functions & Hyperparameter Tuning
Task: Implement and Compare Loss Functions
1.	Define true values (y_true) and model predictions (y_pred).
2.	Compute Mean Squared Error (MSE) and Categorical Cross-Entropy (CCE) losses.
3.	Modify predictions slightly and check how loss values change.
4.	Plot loss function values using Matplotlib.
Expected Output:
•	Loss values printed for different predictions.
•	Bar chart comparing MSE and Cross-Entropy Loss.


```python
# necessary imports
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
#	Define true values (y_true) and model predictions (y_pred).

# true values
y_true = tf.constant([[1, 0, 1]], dtype=tf.float32)

# original predictions
y_pred1 = tf.constant([[0.8, 0.2, 0.7]], dtype=tf.float32)
```

```python
#	Compute Mean Squared Error (MSE) and Categorical Cross-Entropy (CCE) losses.

# MSE of original predictions
mse1 = tf.keras.losses.MeanSquaredError()(y_true, y_pred1).numpy()

# name the CategoricalCrossentropy function
cce = tf.keras.losses.CategoricalCrossentropy()
# CCE of original predictions
cce1 = cce(y_true, y_pred1).numpy()

# print loss values for original predictions
print("Prediction 1 - MSE:", mse1, "CCE:", cce1)
```

```python
#	Modify predictions slightly and check how loss values change.

# new predictions
y_pred2 = tf.constant([[0.7, 0.5, 0.8]], dtype=tf.float32)

# MSE of new predictions
mse2 = tf.keras.losses.MeanSquaredError()(y_true, y_pred2).numpy()

# CCE of new predictions
cce2 = cce(y_true, y_pred2).numpy()

# print loss values for original predictions
print("Prediction 2 - MSE:", mse2, "CCE:", cce2)

# loss values increase due to the decrease in accuracy of predictions. Specifically the second value which changed from .2 -> .5
```

```python
#	Plot loss function values using Matplotlib.

# x-axis labels
labels = ['MSE (Pred1)', 'CCE (Pred1)', 'MSE (Pred2)', 'CCE (Pred2)']
# bar values
values = [mse1, cce1, mse2, cce2]

# initialize matplotlib figure
plt.figure(figsize=(8, 5))
# create a bar plot with specified colors
plt.bar(labels, values, color=['skyblue', 'salmon', 'skyblue', 'salmon'])
# create plot title
plt.title("Comparison of MSE and CCE for Two Predictions")
# create y-axis label
plt.ylabel("Loss Value")
# fit the plot in the figure
plt.tight_layout()
# display plot
plt.show()
# plt.savefig("loss_chart.png")
```
